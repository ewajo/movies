## II. WYKRES - średni przychód i budżet filmu 2010-2016

### 1. Pogrupuj tabelę w taki sposób, aby otrzymać średni przychód (revenue)

```python
df_rev= df[(df['release_date'] >='2010') & (df['release_date'] <= '2016')].groupby('release_date').revenue.sum()

df_2010=round(df[(df['release_date'] >='2010-01-01') & (df['release_date'] <= '2010-12-31')].revenue.mean(),2)
df_2011=round(df[(df['release_date'] >='2011-01-01') & (df['release_date'] <= '2011-12-31')].revenue.mean(),2)
df_2012=round(df[(df['release_date'] >='2012-01-01') & (df['release_date'] <= '2012-12-31')].revenue.mean(),2)
df_2013=round(df[(df['release_date'] >='2013-01-01') & (df['release_date'] <= '2013-12-31')].revenue.mean(),2)
df_2014=round(df[(df['release_date'] >='2014-01-01') & (df['release_date'] <= '2014-12-31')].revenue.mean(),2)
df_2015=round(df[(df['release_date'] >='2015-01-01') & (df['release_date'] <= '2015-12-31')].revenue.mean(),2)
df_2016=round(df[(df['release_date'] >='2016-01-01') & (df['release_date'] <= '2016-12-31')].revenue.mean(),2)

labels=['2010','2011','2012','2013','2014','2015','2016']
data=(df_2010,df_2011,df_2012,df_2013,df_2014,df_2015,df_2016)
df_revenues=pd.DataFrame(data,labels)
data_n=df_revenues.astype(int)
df_revenues_num=pd.DataFrame(data_n)
df_revenues_pl =df_revenues_num.rename(columns= {0:'Średni przychód'}, inplace=True)
df_revenues_pl_rok=df_revenues_num.rename(columns={"":"Rok"},inplace=True)
df_przychod=pd.DataFrame(df_revenues_pl_rok)
df_revenues_num
```

### 2. Pogrupuj tabelę w taki sposób, aby otrzymać średni budżet (budget)
